```python
import torchvision
import matplotlib.pyplot as plt
from torchvision import transforms
from MemSE import ROOT, ProtoDataset
from MemSE.dataset import NORM_COEFS

transform_test = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize(*NORM_COEFS['ImageNet']),
])
dataset = torchvision.datasets.CIFAR10(root=f'{ROOT}/data', train=True, transform=transform_test)
proto = ProtoDataset(dataset)
```

```python
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
for m, c, t in proto:
    plt.imshow(m.permute(1, 2, 0).numpy())
    plt.title(classes[t])
    plt.show()
```
