```python
import torch
import matplotlib.pyplot as plt
from MemSE import ROOT

save_path = ROOT / 'experiments/conference_2/results'
sigmoid_acc_pred = torch.load(save_path / f'trained_SigmoidAccuracyPredictor.pth', map_location=torch.device('cpu'))
acc_pred = torch.load(save_path / f'trained_AccuracyPredictor.pth', map_location=torch.device('cpu'))
print(acc_pred.keys())
```

```python
fig = plt.figure()

plt.plot(acc_pred['train_loss'], label='AccPred - train loss')
plt.plot(acc_pred['val_loss'], label='AccPred - val loss')
plt.plot(sigmoid_acc_pred['train_loss'], label='SAccPred - train loss')
plt.plot(sigmoid_acc_pred['val_loss'], label='SAccPred - val loss')

plt.yscale('log')
plt.legend()
plt.show()
plt.close()
```

```python
fig = plt.figure()

plt.scatter(acc_pred['pow_ref'], acc_pred['pow_pred'], label='AccPred')
plt.scatter(sigmoid_acc_pred['pow_ref'], sigmoid_acc_pred['pow_pred'], label='SAccPred')

plt.title('Power correlation')
plt.xlabel('Reference')
plt.ylabel('Predicted')
plt.legend()
plt.show()
plt.close()
```

```python
fig = plt.figure()

plt.scatter(acc_pred['prec_ref'], acc_pred['prec_pred'], label='AccPred')
plt.scatter(sigmoid_acc_pred['prec_ref'], sigmoid_acc_pred['prec_pred'], label='SAccPred')

plt.title('Precision correlation')
plt.xlabel('Reference')
plt.ylabel('Predicted')
plt.legend()
plt.show()
plt.close()
```

```python
fig = plt.figure()

plt.scatter(acc_pred['pow_ref'], acc_pred['prec_ref'], label='Ref')
plt.scatter(acc_pred['pow_pred'], acc_pred['prec_pred'], label='Pred')

plt.title('AccuracyPredictor')
plt.xlabel('Precision')
plt.ylabel('Power')
plt.legend()
plt.show()
plt.close()
```

```python
fig = plt.figure()

plt.scatter(sigmoid_acc_pred['pow_ref'], sigmoid_acc_pred['prec_ref'], label='Ref')
plt.scatter(sigmoid_acc_pred['pow_pred'], sigmoid_acc_pred['prec_pred'], label='Pred')

plt.title('SigmoidAccuracyPredictor')
plt.xlabel('Precision')
plt.ylabel('Power')
plt.legend()
plt.show()
plt.close()
```
